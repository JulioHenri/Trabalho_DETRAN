# Contexto

O presente estudo faz parte de um trabalho que tem por finalidade apresentar a análise estatística para os dados da OPERAÇÃO VERÃO, que faz parte do Programa Estadual de Segurança Viária do DETRAN/PA, realizada no mês de julho de 2021 pela Coordenadoria de Educação de Trânsito.

Esse Programa realiza atividades em municípios do Estado onde se tem o maior fluxo de pessoas e que também registram sinistros de trânsito com morbimortalidade. A coleta de dados foi realizada no mês de julho/2021 por este mês ser considerado, tradicionalmente e culturalmente, o grande momento do VERÃO
Paraense em virtude das férias escolares nos municípios ocorrendo os grandes fluxos de veículos e concentrações de pessoas em localidades de veraneio como por exemplo Salinópolis, Santarém, Bragança, Marapanim, Distrito de Marudá, Distrito de Mosqueiro e nas BR's e PA's que dão acesso a esses Municípios.


image.png

Foi recebido da equipe do Detran quatro arquivos do Word contendo os questionários aplicados, com a opção correta das questões de múltipla escolha e em uma planilha do programa Excel a digitação dos dados obtidos nos formulários
aplicados, separados por formulário e pela localidade onde os dados foram coletados.

Cada questionário era composto de quatro (4) perguntas sobre o perfil do condutor e cinco (5) questões de múltipla escolha sobre segurança e sinalização no trânsito. Os questionários foram aplicados nas localidades de Salinópolis, Benevides, Bragança e Distrito de Mosqueiro.

A equipe do Detran solicitou a retirada dos dados obtidos com a aplicação do
questionário 4 considerando que esses dados foram obtidos apenas no município de
Salinópolis e poderia inserir um viés no resultado, pois os demais questionários foram aplicados nas quatro localidades pesquisadas. A amostra reduziu de 313 para 280 condutores. Finalizou-se essa etapa incluindo a variável localidade para identificar o local da pesquisa.

A particularidade é que os três questionários não eram iguais, mas as
perguntas do perfil eram as mesmas e nas perguntas de múltipla escolha, havia itens comuns. Então, na primeira etapa foi feita a equalização dos questionários para criação do banco de dados para análise. As quinze perguntas relativas ao trânsito provenientes dos três questionários
foram reduzidas na equalização para dez perguntas distintas compostas por seis itens de segurança e quatro de sinalização descritos a seguir.

## Adicionando o gabarito ao df

```python
#perguntas com resposta A
def correcao(df):
  if df['Q3'] == 'SI':
    return 'SI'
  elif df['Q3'] == 'A':
    return 'correto'
  else:
    return 'incorreto'
df['Q3_gab'] = df.apply(correcao, axis = 1)
```

```python
#perguntas com resposta B
for i in df[['Q1','Q2','Q4','Q5','Q8']]:
  def correcao(df):
    if df[i] == 'SI':
      return 'SI'
    elif df[i] == 'B':
      return 'correto'
    else:
      return 'incorreto'
  df[str(i)+'_gab'] = df.apply(correcao, axis = 1)
```

```python
#perguntas com resposta C
for i in df[['Q6','Q7']]:
  def correcao(df):
    if df[i] == 'SI':
      return 'SI'
    elif df[i] == 'C':
      return 'correto'
    else:
      return 'incorreto'
  df[str(i)+'_gab'] = df.apply(correcao, axis = 1)
```

```python
#perguntas com resposta D
for i in df[['Q9','Q10']]:
  def correcao(df):
    if df[i] == 'SI':
      return 'SI'
    elif df[i] == 'D':
      return 'correto'
    else:
      return 'incorreto'
  df[str(i)+'_gab'] = df.apply(correcao, axis = 1)
```

### Sexo X acerto ou erro nas questões

```python
for i in df[['Q1_gab','Q2_gab','Q3_gab','Q4_gab','Q5_gab','Q6_gab','Q7_gab','Q8_gab','Q9_gab','Q10_gab']]:
  contigency = pd.crosstab(df['SEXO'], df[i]) 
  estatistica, p_value, gl, expected = chi2_contingency(contigency[['correto','incorreto']])
  plt.figure(figsize=(12,8)) 
  sns.heatmap(contigency[['correto','incorreto']], annot=True, cmap='coolwarm'); 


  if p_value < 0.05:
    print('Como p-valor da {} = {:.3}, há evidências suficientes para rejeitar a hipótese nula ao nível de significância de 5%'.format(df[i].name, p_value))
  else:
    print('Como p-valor da {}= {:.3}, não há evidências suficientes para rejeitar a hipótese nula ao nível de significância de 5%'.format(df[i].name, p_value))
```
